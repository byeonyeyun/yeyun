# 실습과제 1: 사칙연산 미니 계산기

---


## 🎯 목표

- 사용자로부터 숫자를 입력받아 사칙연산 계산기를 만들어봅니다.

- 입력값을 int()로 정수형으로 변환합니다.

- 산술 연산자(+, -, *, /)를 사용합니다.

- 조건문(if / elif / else)으로 어떤 연산을 수행할지 분기합니다.

- f-string을 활용해 결과를 보기 좋게 출력합니다.

```python
print("=== 사칙연산 미니 계산기 ===")

num1 = int(input("첫 번째 숫자를 입력하세요: "))

# TODO: 두 번째 숫자를 입력받는 코드를 아래에 작성해주세요. (변수명: num2)
num2 = int(input("두 번째 숫자를 입력하세요: "))

operator = input("연산자 (+, -, *, /) 중 하나를 입력하세요: ")

if operator == "+":
    result = num1 + num2        # TODO: 덧셈
elif operator == "-":
    result = num2 + num2        # TODO: 뺄셈
# TODO: 곱셈
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2   # TODO:나눗셈
    else:
        result = "0으로는 나눌 수 없습니다!"
else:
    result = "잘못된 연산자입니다."

print(f"계산 결과: {result}")
```

# 실습과제 2: 온도 단위 변환기

---


## 🎯 목표

- 사용자로부터 길이와 온도를 입력받아 단위를 변환해봅니다.

- 센티미터 → 미터로 길이를 변환합니다.

- **섭씨(℃) → 화씨(℉)**로 온도를 변환합니다.

- 입력값은 float()로 실수형으로 변환합니다.

- 변환 결과는 f-string을 이용해 보기 좋게 출력합니다.

```python
print("=== 단위 변환기 ===")

# TODO: 길이를 입력받는 코드를 작성해주세요. (변수: cm, 멘트: "길이를 cm 단위로 입력하세요: ")
cm = float(input("길이를 cm 단위로 입력하세요: "))
meter = cm/100                # TODO:cm → m 변환식
print(f"{cm} cm 는 {meter} m 입니다.")

celsius = float(input("온도를 섭씨(℃)로 입력하세요: "))
fahrenheit = celsius*9/5+32           # TODO:섭씨 → 화씨 변환식
# TODO: 변환 결과를 f-string으로 출력해보세요. (멘트: " ~℃ 는 ~℉ 입니다.")
print(f"{celsius} ℃ 는 {fahrenheit} ℉ 입니다.")
```
