## 실습 과제 — 환자 관리 통합 시나리오
### 요구사항
1) 환자 대기열 관리 (list 활용)
- 새로운 환자를 입력받아 대기열에 등록한다.
- 응급환자를 입력받아 대기열 맨 앞에 삽입한다.
- 특정 환자가 퇴원하면 대기열에서 삭제한다.
- 남아 있는 환자들을 알파벳 순으로 정렬한다.
- 최종 대기열을 출력한다.

2) 환자 건강 정보 기록 (dict 활용)
- 사용자에게 환자의 키(cm) 와 몸무게(kg) 를 입력받아 딕셔너리에 저장한다.
- 환자의 혈압을 입력받아 딕셔너리에 추가한다.
- 환자의 체중 정보를 꺼내면서 삭제한다.
- 최종 건강 기록과, 삭제된 체중 값을 출력한다.
아래 빈칸을 채워 **모든 assert** 를 통과시키세요.


```python
# 🏥 환자 관리 통합 실습

# 1) 환자 대기열 관리 -----------------
queue = []

# 환자 등록
name = input( "등록할 환자 이름을 입력해주세요:")
queue.append(name)

# 응급환자 맨 앞 삽입
ER = input( "응급환자 이름을 입력해주세요:")
queue.insert(0,ER)

# 특정 환자 퇴원
discharge = input( "퇴원할 환자 이름을 입력해주세요:")
queue.remove(discharge)

# 알파벳 순 정렬
queue.sort()

# 최종 대열 출력
print("최종 대기열:", queue)


# 2) 건강 정보 기록 -----------------
record = {}

# 키와 몸무게 입력 받아 저장
h = input("신장(키)를 입력해주세요(단위:cm)")
w = input("체중를 입력해주세요(단위:kg)")
record = {
    "신장(키)": h,
    "체중": w
}

# 혈압 입력 받아 저장
BP = input("혈압을 입력해주세요:")
record.update({"혈압" : BP})

# 체중 데이터 꺼내기
removed = record.pop("체중")

# 최종 건강 기록, 삭제된 체중 값을 출력
print("최종 건강 기록:", record)
print("삭제된 체중 값:", removed)
```
