## 문제 1) 리스트 기본 메서드
- `append, remove, sort, reverse, insert, pop, clear, extend` 를 사용해 요구사항을 만족하세요.


```python
fruits = ["apple", "banana", "cherry"]
fruits.__class__  # 타입 힌트용 무해한 표현

# 1) 맨 뒤에 "orange" 추가
fruits.append("orange")
# 2) "banana" 삭제 (처음 하나)
fruits.remove("banana")
# 3) 인덱스 1 위치에 "grape" 삽입
fruits.insert(1,"grape")
# 4) 더 많은 과일(kiwi, mango) 이어 붙이기
more = ["kiwi", "mango"]
fruits.extend(more)
# 5) 알파벳 오름차순 정렬
fruits.sort()
# 6) 순서 뒤집기
fruits.reverse()
# 7) 인덱스 0 요소를 꺼내어 변수 first_out 에 저장
first_out = fruits.pop(0)

assert first_out is not None and isinstance(fruits, list)
assert set(fruits).issubset({"apple","cherry","grape","kiwi","mango","orange"})
print("✅ 문제 1 통과 | fruits:", fruits, "| first_out:", first_out)
```

## 문제 2) 튜플 메서드
- `count`, `index` 를 사용하세요.


```python
t = (1, 2, 3, 2, 2)
cnt = t.count(2)
idx = t.index(3)
assert cnt == 3 and idx == 2
print("✅ 문제 2 통과")
```

## 문제 3) 딕셔너리 메서드
- `get, keys, values, items, pop, update, clear` 를 사용하세요.


```python
person = {"name":"Alice", "age":25}

nm = person.get("name")
age_default = person.get("age",)   # 기본값 활용
unknown = person.get("address", "Unknown")

ks = list(person.keys())
vs = list(person.values())
it = list(person.items())

popped_age = person.pop("age")
person.update({"age":26, "city":"Seoul"})

assert nm == "Alice" and age_default == 25 and unknown == "Unknown"
assert set(ks) <= {"name","age","city"}
assert isinstance(it, list)
print("✅ 문제 3-1 통과 | person:", person)

# 모두 삭제
person.clear()
assert person == {}
print("✅ 문제 3-2 통과 | cleared person:", person)
```

## 문제 4) 집합(Set) 메서드/연산
- `add, remove, discard, clear, union, intersection, difference` 를 사용하세요.


```python
a = {1,2,3}
b = {3,4,5}

a.add(10)       # add
u = a.union(b)    # union
i = a.intersection(b)    # intersection
d = a.difference(b)    # difference (a-b)

a.remove(2)        # remove 2
a.discard(99)       # discard 99 (에러 없어야 함)

assert u == {1,2,3,4,5,10}
assert i == {3}
assert d.issuperset({1,10}) and 3 not in d
print("✅ 문제 4-1 통과 | a:", a)

a.clear()         # clear
assert a == set()
print("✅ 문제 4-2 통과 | cleared a:", a)
```

## 문제 5) 자료형 변환 (Casting)
- 예시에 맞는 변환을 채우세요.


```python
c1 = list("ABC")            # ['A','B','C']
c2 = tuple([1,2,3])           # (1,2,3)
c3 = set([1,1,2,2])         # {1,2}
c4 = dict([("a",1),("b",2)]) # {'a':1,'b':2}

assert c1 == ['A','B','C'] and c2 == (1,2,3) and c3 == {1,2} and c4 == {'a':1,'b':2}
print("✅ 문제 5 통과")
```

## 문제 6) `is` vs `==`
- 동일 객체 비교(`is`)와 값 비교(`==`)를 구분하세요.


```python
lst1 = [1,2,3]
lst2 = [1,2,3]
lst3 = lst1

same_value = (lst1 == lst2)
same_object = (lst1 is lst3)

assert same_value is True and same_object is True
print("✅ 문제 6 통과")
```

## 문제 7) `in` / `not in`
- 포함 여부를 판단하는 코드를 채우세요.


```python
fruits = ["apple","banana","cherry"]
text = "Python Programming"
s = {1,2,3}
d = {"name":"Alice","age":25}

c1 = ("apple" in fruits)
c2 = ("Java" not in text)
c3 = (2 in s)
c4 = ("name" in d)      # dict는 키 기준
c5 = ("Alice" in d.values())

assert [c1,c2,c3,c4,c5] == [True, True, True, True, True]
print("✅ 문제 7 통과")
```

## 문제 8) 시퀀스 슬라이싱 복습
- 리스트 슬라이싱을 완성하세요: [start:stop:step]


```python
nums = [0,1,2,3,4,5,6]
a = nums[2:5]      # [2,3,4]
b = nums[::2]  # [0,2,4,6]
c = nums[::-1]          # [6,5,4,3,2,1,0]

assert a == [2,3,4] and b == [0,2,4,6] and c == [6,5,4,3,2,1,0]
print("✅ 문제 8 통과")
```
