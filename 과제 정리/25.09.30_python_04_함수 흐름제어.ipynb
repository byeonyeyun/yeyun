## 문제 4) `*args` — 전부 숫자인가? 

```python
# TODO: all_digits(*texts) → 전부 숫자 문자열이면 True
def all_digits(*texts):
    return all(map(str.isdigit, texts))

assert all_digits("123", "007") is True
assert all_digits("12a", "34") is False
assert all_digits() is True   # 공집합에 대한 all은 True (수학적 관례)
print("✅ 문제 4 통과")
```

## 문제 5) `**kwargs` — 정렬된 '키=값' CSV 문자열

```python
# TODO: join_kv(**info) 구현
# 요구사항:
# - 키를 알파벳순으로 정렬한 뒤 "키=값" 형태로 만들어 콤마(,)로 이어붙여라
# - 예) join_kv(b=2, a=1) -> "a=1,b=2"
# - 힌트: sorted(info) 는 정렬된 키 목록을 돌려준다
def join_kv(**info):
    return ",".join(map(lambda k: f"{k}={info[k]}", sorted(info)))

# 테스트
assert join_kv(b=2, a=1) == "a=1,b=2"
assert join_kv() == ""
print("✅ 문제 5 통과")
```

## 문제 6) Docstring 직접 작성 — 정규화 함수

```python
# TODO: normalize_score(x, *, base=100) 구현
# - 반환: x/base * 100  (정수 비율 %)
# - Docstring에 '정규화' 단어 포함
def normalize_score(x, *, base=100):
    """
    정규화
    """
    return x/base * 100

doc = normalize_score.__doc__ or "정규화"
assert "정규화" in doc
assert normalize_score(80) == 80      # 80/100 → 80%
assert normalize_score(50, base=50) == 100  # 50/50 → 100%
print("✅ 문제 6 통과")
```

## 문제 12) map / filter / zip — 변형 연습

```python
# map: 문자열을 대문자로 변환
words = ["ai", "lab", "Ko"]
uppered = list(map(str.upper, words))  # str.upper
assert uppered == ["AI","LAB","KO"]

# filter: 길이 >= 3인 문자열만
kept = list(filter(lambda w : len(w) >= 3, words))  # lambda w: len(w) >= 3
assert kept == ["lab"]

# zip: 길이가 다른 리스트 묶어 dict로 (겹치는 부분만)
names = ["A","B","C"]
ages = [10, 20]
paired = dict(zip(names, ages))
assert paired == {"A":10, "B":20}
print("✅ 문제 12 통과")
```

## 문제 13) sorted — 길이 기준 정렬 + 마지막 문자 기준 정렬

```python
words = ["bb", "a", "ccc", "aa"]
by_len = sorted(words, key=len)           # len
by_last = sorted(words, key=lambda w : w[-1])          # lambda w: w[-1]

assert by_len == ["a","bb","aa","ccc"] or by_len == ["a","aa","bb","ccc"]
assert by_last == ["a","aa","bb","ccc"]
print("✅ 문제 13 통과")
```

## 문제 14) any / all — 조건 검사

```python
data = [3, -1, 7, 0]
has_neg = any(x < 0 for x in data)  # any
all_le7 = all(x <= 7 for x in data) # all

assert has_neg is True and all_le7 is True
print("✅ 문제 14 통과")
```
