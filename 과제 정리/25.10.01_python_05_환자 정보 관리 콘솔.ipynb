```python
## 실습: 환자 정보 관리 콘솔

# 1. 환자 정보 입력 → dict 만들어 리스트에 추가
# 2. 환자 목록 보기 → for와 enumerate(start=1)로 순번과 함께 출력
# 3. 이름으로 검색 (부분 일치) → for + 조건
# 4. 환자 정보 수정 (메모만 간단히) → 선택한 환자의 memo 업데이트
# 5. 환자 삭제 → 선택한 번호의 환자 제거
# 6. 간단 통계
#  - 총 환자 수 (len)
#  - 평균 나이 (sum / len, 환자 없으면 0)
#  - 성별 분포 (count)
# 7. 데이터 초기화
# 8. 종료 → break
```

```python
patients = []
while True:
    cmd = int(input("메뉴 번호 선택:"))

    if cmd == 1:
        print("환자 정보 입력 창입니다. 환자 정보를 입력하세요.")
        name = input("환자의 이름을 입력하세요:")
        age = int(input("환자의 나이을 입력하세요:"))
        patients.append({
            "name": name,
            "age": age,
        })
    elif cmd == 2:
        print("환자의 목록입니다.")
    elif cmd == 3:
        print("환자 이름 검색 창 입니다. 환자의 이름을 입력해주세요")
        search_name = input("환자의 이름을 입력하세요:")
        matches = [f'이름: {patient["name"]}, 나이: {patient["age"]} 'patient for patient in patients if search_name in patient.get("name")]
        if not matches:
            print("검색 결과가 없습니다.")
        else:
            print(matches)
```
