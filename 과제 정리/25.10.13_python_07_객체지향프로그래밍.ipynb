
## 1) 데코레이터 없이 동작 이해하기 (문법 펼치기)

아래 `log_decorator`와 `greet`가 있습니다.  
`@log_decorator`를 **사용하지 말고**, 동일한 효과가 나도록 코드를 작성하세요.


```python

def log_decorator(func):
    def wrapper():
        print("[LOG] 함수 시작")
        func()
        print("[LOG] 함수 끝")
    return wrapper

def greet():
    print("안녕하세요!")

# TODO: 데코레이터 문법(@...) 대신 함수 대입으로 동일 효과 만들기
# 예: greet = ____________  # <-- 빈칸 채우기용 예시: 풀이 시 이 줄만 남기고 위는 주석/수정해도 됨

greet = log_decorator(greet)
# 테스트 (시각 확인)
greet()
```


## 3) 캡슐화 + `@property`

요구사항
- MapleCharacter 클래스 만들기
- __hp는 외부에서 직접 접근 불가
- get_hp(), set_hp() 메서드로 HP를 안전하게 관리
- @property / @hp.setter 로 같은 기능을 속성 접근 형태로도 구현

```python

class MapleCharacter:
    def __init__(self, name):
        self.name = name
        self.__hp = 100   # private 변수

    # ✅ 1️⃣ getter
    def get_hp(self):
        return self.__hp

    # ✅ 2️⃣ setter
    def set_hp(self, value):
        if value < 0:
            self.__hp = 0
        elif value > 1000:
            self.__hp = 1000
        else:
            self.__hp = value

    # ✅ 3️⃣ property getter
    @property
    def hp(self):
        return self.__hp

    # ✅ 4️⃣ property setter
    @hp.setter
    def hp(self, value):
        if value < 0:
            self.__hp = 0
        elif value > 1000:
            self.__hp = 1000
        else:
            self.__hp = value


# 🧪 테스트 (assert 자동 채점)
m = MapleCharacter("루카")

# ① getter/setter 테스트
m.set_hp(1200)
assert m.get_hp() == 1000, "❌ set_hp(1200) → HP는 1000이어야 합니다"

m.set_hp(-10)
assert m.get_hp() == 0, "❌ set_hp(-10) → HP는 0이어야 합니다"

m.set_hp(500)
assert m.get_hp() == 500, "❌ set_hp(500) → HP는 500이어야 합니다"

# ② property 테스트
m.hp = 2000
assert m.hp == 1000, "❌ m.hp=2000 → HP는 1000이어야 합니다"

m.hp = -50
assert m.hp == 0, "❌ m.hp=-50 → HP는 0이어야 합니다"

m.hp = 750
assert m.hp == 750, "❌ m.hp=750 → HP는 750이어야 합니다"

print("✅ 3) 통과")
```
