```python
# 1. 캐릭터 생성 : 사용자가 이름을 입력하고. 주사위 굴려(random) 능력치를 결정함, 이름은 list로 받기
# 2. 이름 중복 방지 : 이름이 존재하면 새 이름 요청함
# 3. 주사위 능력치 생성 
# 4. 능력치 재굴리기 : 생성 전 Y/N 입력으로 재시도 가능
# 5. 캐릭터 등록 : 생성 완료 시 리스트(Characters)에 저장
# 6. 캐릭터 목록 보기 : 지금까지 만든 캐릭터 이름 및 능력치를 출력
# 7. 종료 : 프로그램을 종료 (while 루프 종료)
```

```python
import random
class Character:
    def __init__(self, name, STR, LUK, DEX, INT):
        self.name = name
        self.STR = STR
        self.LUK = LUK
        self.DEX = DEX
        self.INT = INT
        
    @staticmethod
    def roll_dice():
        return random.randint(4, 8), random.randint(4, 8), random.randint(4, 8), random.randint(4, 8)

Characters=[]

while True:
    print('========== 메이플 캐릭터 만들기 ==========')
    print('1) 캐릭터 생성 및 능력치 생성')
    print('2) 캐릭터 목록 보기')
    print('3) 종료')
    choice = input('번호를 입력하세요:')

    if choice == '1':
        name = input('캐릭터 명을 입력하세요: ')
        while True:
            print('주사위를 굴리세요!')
            STR, LUK, DEX, INT = Character.roll_dice()
            print(f'STR: {STR}, LUK: {LUK}, DEX: {DEX}, INT: {INT}')
            again = input('재시도 하시겠습니까? (Y/N)').upper()
            if again == 'Y':
                continue
            elif again == 'N':
                new_ch =  {
                    "name" : name,
                    "STR" : STR,
                    "LUK" : LUK,
                    "DEX" : DEX,
                    "INT" : INT
                }
                Characters.append(new_ch)
                print(f'캐릭터 {name}이 생성되었습니다.')
            else:
                print("잘못된 입력입니다.")
            break
        
    elif choice == '2':
        print('====캐릭터 목록 보기====')
        for i,ch in enumerate(Characters,start=1):
            print(f"{i}번 이름:{ch.get('name')}. STR:{ch.get('STR')}, LUK:{ch.get('LUK')}, DEX:{ch.get('DEX')}, INT: {ch.get('INT')}")
            
    elif choice == '3':
        end = input("정말 프로그램을 종료하시겠습니까? (Y/N)").upper()
        if end == 'Y':
            break
        elif end == 'N':
            continue
        else:
            print("잘못된 입력입니다..")
```
