## 4. 결측치 처리
### Q4-1. `age` 결측치를 평균으로 채우세요 (`fillna`) - 원본 수정
### Q4-2. `embarked` 결측치를 최빈값으로 채우세요 (`mode()[0]`) - 원본 수정
### Q4-3. `age`가 결측인 행을 제거하세요 (`dropna(subset=[...])`) 대신, 여기서는 이미 Q4-1에서 채웠다면 예시로 `fare`의 결측이 있다면 제거해보세요

```python
# Q4-1 TODO: age 평균으로 채우기
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
titanic
```

```python
# Q4-2 TODO: embarked 최빈값으로 채우기
titanic['embarked'].fillna(titanic['embarked'].mode(),inplace=True)
titanic
```

```python
# Q4-3 TODO: 특정 컬럼 결측치 행 제거
titanic['age'].dropna()
```
