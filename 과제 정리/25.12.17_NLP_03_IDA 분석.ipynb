### 데이터 불러오기 및 전처리(형태소 추출)

```python
import gensim
import gensim.corpora as corpora
from gensim.corpora import Dictionary
from gensim.models.coherencemodel import CoherenceModel
from gensim.models.ldamodel import LdaModel
import pyLDAvis
import pandas as pd
```

```python
# 엑셀 파일을 데이터프레임 형태로 불러오기: 분석할 뉴스 데이터를 불러옵니다.
data = pd.read_excel('/content/정제파일_ver2.xlsx')# 연구자가 설정하는 곳
data = data.iloc[:300]
```

```python
column_name='text'# 연구자가 설정하는 곳
```

### 사용자 지정 단어 삭제

```python
data = data.astype('str')
data=data[column_name]

# 제거할 단어 목록 정의 (필요에 따라 추가 가능)
words_to_remove = ['안녕', '하세', '감사','상담','선생님','의사','말씀','건강','계세']  # 여기에 제거하고 싶은 단어들을 추가하세요
```

```python

# 정확한 단어 일치로 제거하기 위한 함수
def remove_exact_words(text, words_list):
    if isinstance(text, str):
        # 공백으로 분리하여 단어 단위로 처리
        words = text.split()
        # 제거할 단어 목록에 없는 단어만 유지
        filtered_words = [word for word in words if word not in words_list]
        # 다시 문자열로 결합
        return ' '.join(filtered_words)
    return text
# 텍스트에서 지정한 단어들 제거 (Series에 직접 적용)
data = data.apply(lambda x: remove_exact_words(x, words_to_remove))
```

### LDA 분석

```python
# 1️⃣ 텍스트 데이터를 단어 단위로 분할 (토큰화)
# 'data' 리스트에 있는 텍스트(예: '제목_본문')를 공백 기준으로 나누어 단어 리스트(token) 생성
token = [keywords.split() for keywords in data]

# 2️⃣ 단어별 고유 ID를 부여하는 사전(Dictionary) 생성
# Gensim의 Dictionary 객체를 사용하여 각 단어에 ID를 할당
# 결과적으로, 모든 단어가 ID와 함께 저장됨
id2word = corpora.Dictionary(token)

# 3️⃣ 너무 적거나 너무 많은 문서에서 등장하는 단어 제거
# no_below=5 → 최소 5개 문서 이상에서 등장한 단어만 유지 (너무 희귀한 단어 제거)
# no_above=0.2 → 전체 문서의 20% 이상에서 등장하는 단어 제거 (너무 흔한 단어 제거)
id2word.filter_extremes(no_below=2, no_above=0.2)

# 4️⃣ 문서를 숫자로 변환 (LDA 모델 학습을 위한 형태)
# 각 문서를 (단어 ID, 해당 단어 등장 횟수) 형태의 리스트로 변환
# LDA 모델이 학습할 수 있는 형태의 문서-단어 행렬(corpus) 생성
corpus = [id2word.doc2bow(text) for text in token]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from gensim.models.ldamodel import LdaModel

# 퍼플렉시티 값을 저장할 빈 리스트를 초기화합니다.
perplexity_values = []

# 토픽의 수를  변화시키면서 LDA 모델의 퍼플렉시티를 계산합니다.
for i in range(2, 15):
    # LDA 모델을 생성합니다.
    ldamodel = LdaModel(corpus, num_topics=i, id2word=id2word, random_state=100)

    # 모델의 퍼플렉시티를 계산합니다.
    perplexity = np.exp(ldamodel.log_perplexity(corpus))

    # 계산된 퍼플렉시티를 리스트에 추가합니다.
    perplexity_values.append(perplexity)

# 그래프에 사용할 x값 설정
x = range(2, 15)

# Perplexity 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(x, perplexity_values, marker='o', color='black', markerfacecolor='black')
plt.xlabel("Number of Topics")
plt.ylabel("Perplexity Score")
plt.title("Perplexity Score by Number of Topics")

# 각 데이터 포인트에 수치값 라벨 추가
for i, value in enumerate(perplexity_values):
    plt.text(x[i], value, f"{value:.5f}", ha='center', va='bottom')

plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig(f'./perplexity.png')
plt.show()
```

```python
 # CoherenceModel
from gensim.models import CoherenceModel

coherence_values=[]
for i in range(2,15):
  ldamodel=gensim.models.ldamodel.LdaModel(corpus, num_topics=i,id2word=id2word,random_state=100)
  coherence_model_lda=CoherenceModel(model=ldamodel, texts=token,dictionary=id2word,topn=10)
  coherence_lda=coherence_model_lda.get_coherence()
  coherence_values.append(coherence_lda)
```

```python
x=range(2,15)
plt.figure(figsize=(10, 6))  # 그래프 크기 조정
plt.plot(x, coherence_values, marker='o', color='black', markerfacecolor='black')  # 선, 마커 색상을 블랙으로 설정
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.title("Coherence Score by Number of Topics")
# 각 데이터 포인트에 수치값 라벨 추가
for i, value in enumerate(coherence_values):
    plt.text(x[i], value, f"{value:.5f}", ha='center', va='bottom')
plt.savefig(f'./coherence.png')
plt.show()
```

```python
##적정토픽수 입력 coherence를 최대로 하는 적절합 토픽수
k = 4 # 연구자가 설정하는 곳
```

```python
#적정토픽에 해당 되는 단어를 구하기!
from gensim import corpora, models
# 적절한 토픽 수를 설정하여 LDA 모델 훈련
model = LdaModel(corpus=corpus, num_topics=k, id2word=id2word, passes=15,random_state=100)  # num_topic에 적절 토픽 수 설정
topics = model.print_topics(num_words=10) ####토픽당 단어수설정
```

```python
# 도출된 토픽을 Excel 파일로 저장
topic_data = {}

for topic_id, topic in topics:
    words_weights = [item.split("*") for item in topic.split(" + ")]
    words = [ww[1].strip('"') for ww in words_weights]
    weights = [float(ww[0]) for ww in words_weights]
    topic_data[f"Topic {topic_id+1} 단어"] = words
    topic_data[f"Topic {topic_id+1} 가중치"] = weights

# topic_data 딕셔너리를 데이터프레임으로 변환
topic_df = pd.DataFrame(topic_data)

# Excel 저장을 위한 writer 객체 생성
with pd.ExcelWriter('lda_results.xlsx', engine='openpyxl') as writer:
    # 상세 정보가 포함된 시트 저장
    topic_df.to_excel(writer, sheet_name='Detailed Topics')

    # 원래 형식의 시트도 저장
    TOPIC = pd.DataFrame({'topics': topics})
    TOPIC.to_excel(writer, sheet_name='Original Format')

print("LDA 결과가 성공적으로 Excel 파일로 저장되었습니다.")
```

```python
# 한글 폰트 경로를 지정합니다. WordCloud에서 한글을 정상적으로 표시하기 위한 준비 단계입니다.
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib.font_manager as fm
import numpy as np

# 워드클라우드에 적용할 폰트 설정
wc_font_path = font_path

# 단어 가중치를 딕셔너리로 변환하는 함수
def get_word_weights_dict(topic_id):
    words = topic_data[f"Topic {topic_id+1} 단어"]
    weights = topic_data[f"Topic {topic_id+1} 가중치"]  # 키 이름 주의
    return {words[i]: weights[i] for i in range(len(words))}

# 토픽 수 확인
num_topics = len(topics)

# 토픽별 색상 테마 - 진한 색상 사용
topic_colormaps = ['Blues_r', 'Oranges_r', 'Greens_r', 'Purples_r']

# 그림 크기 설정 - 충분히 큰 크기로 단순화
plt.figure(figsize=(20, 5))

# 토픽별 서브플롯 생성 - 심플하게
for i in range(num_topics):
    plt.subplot(1, num_topics, i+1)  # 1행 n열 구조로 심플하게

    # 단어-가중치 딕셔너리 생성
    word_weights = get_word_weights_dict(i)

    # 워드클라우드 생성 - 충분히 큰 크기, 심플한 설정
    wordcloud = WordCloud(
        font_path=wc_font_path,
        width=1200,
        height=800,
        background_color='white',
        max_words=80,  # 단어 수 약간 줄임
        colormap=topic_colormaps[i % len(topic_colormaps)],
        prefer_horizontal=0.9,
        relative_scaling=0.6,
        random_state=42
    ).generate_from_frequencies(word_weights)

    # 워드클라우드 표시
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')

    # 제목 설정 - 적당히 큰 크기로 단순화
    plt.title(f'TOPIC {i+1}', fontsize=20, fontweight='bold', pad=20)

# 레이아웃 조정 - 심플하게
plt.tight_layout(pad=2)

# 저장 및 표시
plt.savefig('topic_wordclouds.png', dpi=300, bbox_inches='tight')
plt.show()
```
