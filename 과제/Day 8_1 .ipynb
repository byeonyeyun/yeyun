```python
class Character:                                #기본값을 만들수있음
    server = "크로아"
    user_count = 0
    def __init__(self, name, STR, LUK, DEX, INT, job = "초보자"):
        self.name = name
        self.__STR = STR
        self.__LUK = LUK
        self.__DEX = DEX
        self.__INT = INT
        self.__level = 1
        self.__exp = 0
        self.__hp = 100
        self.__mp = 100
        self.__glod = 0
        self.__stats_point = 0
        Character.user_count += 1
    
    def get__level(self):  #getter
        print(self.__level)

    def get_stats(self): # getter
        print(f'STR:{self.__STR}, LUK:{self.__LUK}, DEX: {self.__DEX}, INT:{self.__INT}')

    def get__exp(self): # 아래 setter로 만든 걸 보여주기 위한 getter
        print(self.__exp)
    
    def gain__exp(self, amount):  # setter
        self.__exp += amount
        if self.__exp >= 100:  # exp가 100올라가면 level 1이 증가됨
            self.__level += 1
            self.__exp -= 100  # 그리고 다시 exp에 -100해서 exp를 0으로 만듬

    def add_str(self,amount): # setter
        if self.__stats_point >= amount :
            self.__STR += amount
            self.__stats_point -= amount
            print(f'힘이 {amount}만큼 증가했습니다.')
        else:
            print(f'스탯포인트가 부족합니다.')
            
    @property
    def hp(self):
        return self.__hp

    @property
    def level(self):
        return self.__level

    @hp.setter
    def hp(self, amount):
        if amount<0:
            self.__hp = 0
        else:
            self.__hp = amount
    
    def info(self):
        print(f'{self.name}님의 레벨은 {self.__level}입니다. {self.server}')

    def attack(self):
        print(f'{self.name}님이 공격을 당했습니다.')
        
# 유저수 측정
    # 인스턴스 메서드
    def total_users(self):   
        print(f'총 유져 수: {self.user_count}')
    # 클래스 메서드
    @classmethod
    def total_users2(cls): 
        print(f'총 유저 수: {cls.user_count}')
    # 정적 메서드
    @staticmethod
    def greet():
        print("안녕하세요")
```

```python
class Warrior(Character):
    def __init__(self, name, STR, LUK, DEX, INT, weapon):
        super().__init__(name, STR, LUK, DEX, INT) # 여기는 self가 빠짐 주의
        self.weapon = weapon

    def attack(self, monster:Monster): #메서드 오버라이딩
        exp = monster.take_damage(10) # 캐릭터와 몬스터 전투 예시
        if exp:           # 전투 후 보상
            self.gain__exp(exp)
            print(f'몬스터를 처치해서 {exp}만큼 경험치가 올랐습니다.')
        return(f'{self.weapon}을 사용하였습니다.')
```

```python
#캐릭터와 몬스터 전투 예시
class Monster:
    def __init__(self, name, exp, hp):
        self.name = name
        self.__exp = exp
        self.__hp = hp
    @property
    def hp(self):
        return self.__hp

    def take_damage(self,dmg):
        self.__hp = max(0, self.__hp - dmg ) # 최소값이 0이 되게
        print(f'{self.name}몬스터가 {dmg}만큼 피해를 입어서 남은 체력이 {self.__hp}입니다.')
        if not self.__hp:     # 전투 후 보상
            return self.__exp
        return 0
```

```python
mushroom = Monster('주황버섯',10,100)
```

```python
warrior1.attack(mushroom)    # 전투 후 보상
```
