```python
import pandas as pd
```

```python
#데이터 프레임 생성 방법
data = {
    "name": ["동원","태규","희주","승재"],
    "age" : [32, 29, 28, 31],
    "score" : [3, 5, 4, 1]
}
df = pd.DataFrame(data)
df
```

```python
df.index = ['환자1', '환자2', '환자3', '환자4']
df
```

```python
# 단일 칼럼(Column) 선택
df['name']
```

```python
# 다중 칼럼(Column) 선택
df[['name','age']]
```

```python
# 행 선택 (loc) : 인덱스 이름 기반으로 선택
df.loc['환자1'] 
```

```python
df.loc[['환자1','환자2']] #다중 인덱스 선택
```

```python
# 행 선택 (iloc) : (정수 기반) 인덱스 행번호 선택
df.iloc[0]
```

```python
df.iloc[::2] #슬라이싱도 가능
```

```python
df.iloc[1::2][['name','age']] #슬라이싱 +컬럼만 추출
```

```python
df[df['age']>30] # 행 필터링
```

```python
df[(df["age"]>30) | (df["score"]>4)]  #여러 조건 결합
```

```python
df.sort_values('age') # 오름차순
```

```python
df.sort_values('age',ascending=False) #내림차순
```

```python
# 값 수정하기
df.loc[df['name'] == '승재', 'score'] =6
df
```

```python
# 컬럼 추가
df["isYoung"] = df['age'] <30
df
```

```python
df
```

```python
# Column(컬럼,열,세로)을 지우고 싶을때
df.drop(columns = ["isYoung"], inplace=True) 
# Columns = [""] 로 쓰거나 axis = 1
df
```

```python
# row(로우,행,가로) 지우고 싶을때
df.drop(index = ['환자2'], inplace=True) 

# index = [""] 로 쓰거나  Axis = 0 
df
```

```python
# 위에 식이랑 같은 것
df = df.drop(columns = ["isYoung"])
df
```

```python
# 행 추가
df.loc['환자2'] = ['태규',29,5]
df
```

```python
df.index = [1, 2, 3, 4]
df
```

```python
df.sort_index()
```
