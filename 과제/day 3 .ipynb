```python
# list
# 일차원 배열
a = ["apple", 0 , 0.1 , True] #list 안에는 다양한 자료형들이 들어감
```

```python
# list에는 시퀀스(명령의 순서)가 있어 index접근이 가능함
a[2] # 2의 위치에 있는 0.1이 나옴
```

```python
# a라는 리스트에 있는 "apple" 또한 시퀀스가 있어 Index 접근이 가능함
a[0][1]
```

```python
# 이차원배열 : list 안에 list를 또 쓸 수 있음
a = ["apple", 0 , 0.1 , True, ['banana', 1, 0.1, False]]
```

```python
a[-1][0][2]
```

```python
#시퀀스가 있기 때문에 콜론 ( : ) 사용가능
a[:-1]
```

```python
#append
fruits = ["apple","orange", "cherry", "banana"]
fruits.append("orange") # 맨뒤에 추가됨
fruits
```

```python
# remove
fruits.remove("orange") # 처음 나온 orange를 삭제
fruits
```

```python
# sort
fruits.sort()
fruits
```

```python
 # sort로 역순으로 정렬시
fruits.sort(reverse=True) 
fruits
```

```python
# reverse
fruits.reverse()
fruits
```

```python
# insert
fruits.insert(2,"grape")
fruits
```

```python
# pop 
fruits.pop(0) # '잘라내기'라고 생각하면됨 , return값이 존재함
# index에 아무 값이 없으면(=생략하면) ex) fruits.pop() 이면 마지막 값을 잘라냄
```

```python
# clear
fruits.clear()
fruits
```

```python
# extend(iterable)
fruits = ["apple", "banana"]
more = ["watermelon", "mango"]
fruits.extend(more)
fruits
```

```python
# 값을 바꿔도 list의 주소값은 안바뀜 : Mutable (가변형) 
# list는 바꿔도 덮어씌여지는게 아니라 요소만 바뀌는것임 (가변형) 
fruits[0] = "pineapple"
fruits
```

```python
# 중괄호
# 아래 세팅 대로 씀 콤마(,) 빼먹지 말기
person = {
    "name": "미후",
    "age": 10,
    "address": "서울시",
    "lauguages": ['korean'],
    123: 123123
}  # key : value
```

```python
person
```

```python
# index 접근은 할 수 있지만 시퀀스하지는 않아서 순서가 아니라 key를 적어야됨
person["name"]
```

```python
# get
person.get("name")
```

```python
# 없는 key입력이 입력되면 아무것도 안나오거나 or 내가 설명한 default값이 나온다 
# 고로 error가 나오지 않아서 안정적이다
person.get(1234,"값이 없습니다.")
```

```python
# keys
person.keys() # 아래 타입을 dict_keys라고 나옴
```

```python
# values
person.values()
```

```python
# items
person.items() #튜플형식으로 나옴
```

```python
# pop
person.pop(123)
```

```python
# update
new_person = {
    "age": 12,
    "city": "Seoul"
}
```

```python
person.update(new_person)
person # 같은 key값은 변경 ,다른 key값은 추가됨
```

```python
person.update({
    "age": 12,
    "city":"Seoul"
})
# 위에와 같음
```

```python
# is 연산자
a = [1, 2, 3]
b = [1, 2, 3]
c = a

a is b 
# a is b는 서로 같은 객체인지를 묻는 질문이 되며, a와 b의 주소값이 다르기 때문에 False가 나옴
#레퍼런스 타입은 메모리 스택에 실제 값이 존재하는게 아니라 메모리 Heap을 가르키는 주소값이 저장되어있음
```

```python
print (a==b) # 값은 같음
print (a is b) #서로 다른 리스트 객체
print (a is c) # 같은 객체 참조
```

```python
# in 연산자
# dict
person = {"name": "Alice", "age" : 25 }
print("name" in person) # key 기준 검사
print("Alice" in person) # 값은 직접 검사안함
# 값을 확인하려면 values() 사용
print("Alice" in person.values())
```
