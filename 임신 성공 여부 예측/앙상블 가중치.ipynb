```python
import pandas as pd

# ===============================
# [STEP 1] Submission 파일 로드
# ===============================
sub_binary = pd.read_csv('feature_5_2hr_drop_binary_autogluon_submission.csv')                
sub_default = pd.read_csv('feature_5_2hr_drop_autogluon_submission.csv')

# 예측 컬럼 자동 탐지 (마지막 컬럼 기준)
target_col = sub_binary.columns[-1]

# 컬럼 일치 여부 확인
assert list(sub_binary.columns) == list(sub_default.columns), "Submission 컬럼 불일치"

# ===============================
# [STEP 2] 단일 가중치(0.45) 앙상블 적용
# ===============================
# Binary 모델에 0.45, Default 모델에 0.55 (1-0.45) 적용
w = 0.45

print(f"앙상블 적용 중... Binary Model: {w} / Default Model: {1-w:.2f}")

final_pred = (
    sub_binary[target_col] * w +
    sub_default[target_col] * (1 - w))

# ===============================
# [STEP 3] 결과 저장
# ===============================
submission = sub_binary.copy()
submission[target_col] = final_pred

filename = f'ensemble_w_{w:.2f}.csv' 
submission.to_csv(filename, index=False)

print(f"🎉 {filename} 파일 생성 완료!")                      
```
